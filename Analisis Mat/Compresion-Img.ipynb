```python
from matplotlib.image import imread
from matplotlib.image import imsave
import matplotlib.pyplot as plt
import numpy as np
import os


file='C:/Users/Dicsys/Downloads/Monedas2.jpg';

img_orig = imread(file); # En img_orig se guarda un arreglo de tres matrices cada una de ellas contiene la intensidad del piexel en los canales RGB.
#A=img_orig[:,:,0]; #Ac se toma uno de los canales.


A=np.mean(img_orig,-1);# Ac se promedia la intensidad de todos los canales.


# Se imprime la imgen.
img=plt.imshow(A);
img.set_cmap('gray');
plt.show();


plt.show();
```

```python
U, S, VT=np.linalg.svd(A) # Se calcula la SVD de la matriz con la que estamos trabajando.
S=np.diag(S) #La SVD devuelve los valores singulares y ac construimos la matriz Sigma.


plt.figure(1)
plt.plot (S)
plt.axis([0,400,0,6000])
plt.show()


# Se seleccina cuntos valores singulares se toman para reconstruir la imgen.
k=50; 
Ared = U[:,:k] @ S[0:k,:k] @ VT[:k,:]; # Recompone la imgen usando los k valores singulares seleccionados.


# Se muestra la imgen reducida.
img=plt.imshow(Ared);
img.set_cmap('gray');
plt.show();



# Guardar la imagen.
file_red=os.path.splitext(file)[0]+'_red'+os.path.splitext(file)[1]; # esto solo es para agregar "_red" al archivo con la imgen reducida
imsave(file_red,Ared,cmap="gray");



# Esto es para comparar los tamaos de los archvos y ver cunto se redujo.
sizeOrig=os.stat(file).st_size;
sizeRed=os.stat(file_red).st_size;
reduc=100*(sizeOrig-sizeRed)/sizeOrig; # Porcentaje de reduccin.
print(f'El porcentaje de reduccin es {reduc:.2f} %');
```
