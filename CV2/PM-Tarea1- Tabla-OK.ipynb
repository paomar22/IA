```python
import torch
from torch import nn
```

```python
# Input values 
input = torch.randn(1, 3, 64, 64)
```

```python
#Conv2D(f=5, s=2, c=8, p=2
conv = nn.Conv2d(in_channels=3, out_channels=8, kernel_size=5, stride=2, padding=2)
out_conv = conv(input)
print("Dimensiones de la activacion:", out_conv.size())
conv_params = sum(p.numel() for p in conv.parameters())
print("Cant Parametros:", conv_params)
```

```python
#Maxpool(f=2, s=2)
maxpool = nn.MaxPool2d(kernel_size=2, stride=2)
out_maxpool = maxpool(out_conv)
print("Dimensiones de la activacion:", out_maxpool.size())
maxpool_params = sum(p.numel() for p in maxpool.parameters())
print("Cant Parametros:", maxpool_params)
```

```python
#Conv2D(f=3, s=1, c=16, p=1)
conv1 = nn.Conv2d(in_channels=8, out_channels=16, kernel_size=3, stride=1, padding=1)
out_conv1 = conv1(out_maxpool)
print("Dimensiones de la activacion:", out_conv1.size())
conv_1_params = sum(p.numel() for p in conv.parameters())
print("Cant Parametros:", conv_1_params)
```

```python
#Maxpool(f=4, s=4)
maxpool1 = nn.MaxPool2d(kernel_size=4, stride=4)
out_maxpool1 = maxpool1(out_conv1)
print("Dimensiones de la activacion:", out_maxpool1.size())
maxpool_1_params = sum(p.numel() for p in maxpool1.parameters())
print("Cant Parametros:", maxpool_1_params)
```

```python
# Dense(salida=6)
linear = nn.Linear(in_features=16 * 4 * 4, out_features=6)
out_linear = linear(out_maxpool1.view(-1, 16 * 4 * 4))
print("Size:", out_linear.size())
linear_total_params = sum(p.numel() for p in linear.parameters())
print("Parameters:", linear_total_params)
```
