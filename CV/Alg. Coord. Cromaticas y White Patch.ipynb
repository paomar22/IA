```python
#Si queremos que las imágenes sean mostradas en una ventana emergente quitar el inline
%matplotlib inline  

# OpenCV-Python utiliza NumPy para el manejo de imágenes
import numpy as np
# cv2 es el módulo python para acceder a OpenCV 
import cv2 as cv
# Usamos las poderosas herramientas de graficación de matplotlib para mostrar imágenes, perfiles, histogramas, etc
import matplotlib.pyplot as plt

from PIL import Image, ImageEnhance
```

```python
def carga_imagen(path):
    img = cv.cvtColor(cv.imread(path), cv.COLOR_BGR2RGB)
    return img

def coordenadas_cromaticas(img):
    im_r,im_g,im_b = cv.split(img)
    im_r=np.uint8(im_r)
    im_g=np.uint8(im_g)
    im_b=np.uint8(im_b)
    suma_canales=cv.add(im_r,cv.add(im_g,im_b))
    new_red=im_r/(suma_canales+0.001)
    new_green=im_g/(suma_canales+0.001)
    new_blue=im_b/(suma_canales+0.001)
    new_image = np.stack([new_red,new_green,new_blue],axis=2)
    return (new_image)

def white_patch(img):

    im_r=np.uint8(img[..., 0])
    im_g=np.uint8(img[..., 1])
    im_b=np.uint8(img[..., 2])
    
    # red
    im_r_max=np.percentile(im_r,100)
    
    # green
    im_g_max=np.percentile(im_g,100)
    
    # blue
    im_b_max=np.percentile(im_b,100)
    
    
    new_red=(255/im_r_max)*im_r
    new_green=(255/im_g_max)*im_g
    new_blue=(255/im_b_max)*im_b
    
    new_image = np.stack([new_red,new_green,new_blue],axis=2).astype(np.uint8)
    return (new_image)
```

```python

img1 = carga_imagen('C:/Users/Paola Mariscal/Downloads/CoordCrom_1.png')
img2 = carga_imagen('C:/Users/Paola Mariscal/Downloads/CoordCrom_2.png')
img3 = carga_imagen('C:/Users/Paola Mariscal/Downloads/CoordCrom_3.png')

new_image1=coordenadas_cromaticas(img1)
new_image2=coordenadas_cromaticas(img2)
new_image3=coordenadas_cromaticas(img3)


fig, axis = plt.subplots(ncols=3,nrows=2,figsize=(18,6))

axis[0][0].imshow(img1)
axis[1][0].imshow(new_image1)

axis[0][1].imshow(img2)
axis[1][1].imshow(new_image2)

axis[0][2].imshow(img3)
axis[1][2].imshow(new_image3)

plt.show()
```
