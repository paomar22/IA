```python
#Si queremos que las imgenes sean mostradas en una ventana emergente quitar el inline
%matplotlib inline  

# OpenCV-Python utiliza NumPy para el manejo de imgenes
import numpy as np
# cv2 es el mdulo python para acceder a OpenCV 
import cv2 
# Usamos las poderosas herramientas de graficacin de matplotlib para mostrar imgenes, perfiles, histogramas, etc
import matplotlib.pyplot as plt


eyes=cv2.imread('C:/Users/Paola Mariscal/Downloads/eyes.jpg')

img_ori=eyes.copy()

img=cv2.cvtColor(eyes, cv2.COLOR_BGR2GRAY)

plt.rcParams["figure.figsize"]=(16,9)
plt.imshow(img,cmap="gray")
```

```python
img = cv2.imread('C:/Users/Paola Mariscal/Downloads/eyes.jpg',0)
img = cv2.medianBlur(img,5)
cimg = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)

circles = cv2.HoughCircles(img ,cv2.HOUGH_GRADIENT,1,50,#70
param1=50,param2=19,minRadius=10,maxRadius=17)

#cv.HoughCircles(processed, storage, cv.CV_HOUGH_GRADIENT, 2, 100.0, 30, 150,100,140)
circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)
    # draw the center of the circle
    cv2.circle(cimg,(i[0],i[1]),2,(0,0,255),3)

#cv2.imshow('detected circles',cimg)
#cv2.waitKey(0)
#cv2.destroyAllWindows()

plt.rcParams["figure.figsize"]=(16,9)
plt.imshow(cimg)
plt.show()
```
